## 1. Row-major order

Numpy biblioteka cuva visedimenzione nizove po vrstama (row major order) kao i C programski jezik. Na primer Fortran ima column major order.

```python
import numpy as np

a=np.array([[1,2,3],[4,5,6]])
b=a*2
print(a,'\n',b)
print(np.concatenate((a,b))) # po default je spajanje axis=0, spajanje po vrstama
print(np.concatenate((a,b),axis=1)) #spajanje po axis=1
```

```python
a=np.array([1,2,3,4,5,6])
print(a)
print(a.reshape((2,3))) #reshape po C order #order='C' default
print(a.reshape((2,3),order='F'))
```

## Korisna numpy metoda meshgrid()

Ova funkcija se koristi da napravi pravougaonu mrezu tacaka na osnovu dva 1D niza koji prate Dekartovo indeksiranje. Meshgrid je inspirisan funkcijom iz MATLAB-a.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, -1, 5)
y = np.linspace(-5, -1, 5)

# Generiemo koordinatne mree pomou meshgrid
X, Y = np.meshgrid(x, y)

# Prikazujemo rezultat
print("X koordinate:")
print(X)
print("\nY koordinate:")
print(Y)

# Prikazujemo take na grafiku
plt.scatter(X, Y)
plt.xlabel('X koordinate')
plt.ylabel('Y koordinate')
plt.title('Koordinatne mree')
plt.show()
```

#### Ove matrice mogu se koristiti kao ulazne vrednosti za funkciju koja zavisi od dve promenljive. Na primer, ako imate funkciju z = f(X, Y), moete evaluirati vrednosti funkcije za svaku taku na mrei tako to ete koristiti matrice X i Y.

```python
import matplotlib.pyplot as plt
from matplotlib import cm

z = np.sin(X**2 + Y**2) / (X**2 + Y**2)
print (z.shape)
print (z)
print(X)
print(Y)

#CRTANJE POMOCU PCOLORMESH
plt.pcolormesh(x,y,z,cmap=cm.jet) #rainbow, viridis
plt.colorbar()
```

Pandas takoe moze da izbrie redove koji nisu relevantni ili sadre pogrene vrednosti, kao to su prazne ili NULL vrednosti. Ovo se zove *icenje podataka*. I redovi i kolone imaju indekse, a operacije moete vriti na redovima ili kolonama odvojeno.
