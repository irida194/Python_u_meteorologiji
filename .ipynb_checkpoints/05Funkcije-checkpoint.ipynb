# Funkcije

Funkcija je kao mini program unutar programa.<br>

Glavna namena funckija je da grupisu kod koji ce biti izvrsen vise puta (podela na logicke grupe). Bez definisane funkcije bi trebalo da kopirate ovaj kod svaki put, a program bi izgledao nepregledno.<br>

### Vracene vrednosti i iskazi return

Kada pozovete funkciju len() i prosledite joj argument kao sto je 'Hello', pozvivanje funkcije se svodi na vrednost celog broja 5, sto je duzina stringa koji ste joj prosledili. Genralno, vrednost koju izracunava poziv funkcije naziva se vracena vrednost funkcije. 

U ovom primeru koristimo ugradjenu funckiju [random](https://docs.python.org/3/library/random.html)

```python
#program magic8Ball.py
import random

def getAnswer(answerNumber):
    if answerNumber == 1:
        return 'It is certain'
    elif answerNumber == 2:
        return 'It is decidedly so'
    elif answerNumber == 3:
        return 'Yes'
    elif answerNumber == 4:
        return 'Reply hazy try again'
    elif answerNumber == 5:
        return 'Ask again later'
    elif answerNumber == 6:
        return 'Concentrate and ask again'
    elif answerNumber == 7:
        return 'My reply is no'
    elif answerNumber == 8:
        return 'Outlook not so good'
    elif answerNumber == 9:
        return 'Very doubtful'
r = random.randint(1,9) #Ova funkcija svodi na nasumicni ceo broj izmedju 1 i 9, ukljucujuci i 9
fortune = getAnswer(r)
print(fortune)
```

U Pythonu ne možete direktno vratiti više vrednosti sa jednom return naredbom. Međutim, možete vratiti više vrednosti kao tuple, a zatim raspakovati vrednosti kada pozivate funkciju. Evo kako to možete uraditi:

```python
def avg_max_temp(temps):
    avgT=sum(temps)/len(temps)
    maxT=max(temps)
    return avgT, maxT

temps=[16.3, 15, 17.8, 20.1, 22, 23.1, 23.7, 18, 18.7]
#avg_max_temp(temps) #vraca kao rezultat tuple
#avgT, maxT = avg_max_temp(temps) #trik visestruke dodele

avgT=avg_max_temp(temps)[0]
maxT=avg_max_temp(temps)[1]
print(round(avgT,3),maxT)
```

U primeru ispod, znak * se koristi kako biste omogućili da funkcija avg_max_temp prima proizvoljan broj argumenata. Kada koristite * ispred parametra funkcije, to označava da taj parametar prima proizvoljan broj argumenata i pretvara ih u tuple.

```python
def avg_max_temp(*temps):
    print(type(temps))
    avgT=sum(temps)/len(temps)
    maxT=max(temps)
    return avgT, maxT

avgT, maxT = avg_max_temp(16.3, 15, 17.8, 20.1, 22, 23.1, 23.7, 18, 18.7) #trik visestruke dodele
print(round(avgT,3),maxT)
```

**REKURZIVNO POZIVANJE FUNKCIJE**

Evo kako funkcija radi:

Ako je n jednako 1, funkcija će vratiti 1. Ovo je bazni slučaj rekurzije, jer je faktorijel broja 1 jednak 1.

U suprotnom, ako n nije jednako 1, funkcija će rekurzivno pozvati samu sebe sa argumentom n - 1 i pomnožiti rezultat tog poziva sa n. To znači da će faktorijel broja n biti jednak n pomnoženo sa faktorijelom broja n - 1.

Ovo će se nastaviti rekurzivno pozivati sve dok ne dođe do baznog slučaja (kada je n jednako 1), a zatim će se vrednosti faktorijela brojeva ugnježdenih poziva rekurzije pomnožiti kako bi se dobila konačna vrednost faktorijela broja n.

Na primer, ako pozovemo faktRekurzija(5), funkcija će izračunati faktorijel broja 5 na sledeći način:

faktRekurzija(5) = 5 * faktRekurzija(4) <br>
faktRekurzija(4) = 4 * faktRekurzija(3) <br>
faktRekurzija(3) = 3 * faktRekurzija(2) <br>
faktRekurzija(2) = 2 * faktRekurzija(1) <br>
faktRekurzija(1) = 1 (bazni slučaj)

```python
#Izbegavati jer je racunski skupo
def faktRekurzija(n):
    if n == 1:
        return 1
    else:
        return n * faktRekurzija(n - 1)

faktRekurzija(5)
```

1. **SPORO IZVRSENJE** :Svaki put kad se funkcija rekurzivno poziva, podaci o pozivu funkcije (stanje) moraju biti smešteni na stek poziva. To uključuje informacije o argumentima i lokalnim promenljivima. Ako rekurzija ide duboko ili ako ima mnogo rekurzivnih poziva, stek poziva može postati veoma dubok, što može dovesti do potrošnje velike količine memorije.

2. **Vreme izvršenja**: Rekurzivna funkcija može biti spora zbog ponovljenih poziva same sebe. Svaki put kad se funkcija pozove rekurzivno, mora se izvršiti nova instanca funkcije, sa svim overheadom povezanim sa pozivom funkcije, uključujući i dodatne operacije poput testiranja uslova za bazni slučaj i prelaza argumenata.

3. **Teškoće sa čitanjem i održavanjem koda**: Rekurzivni kod često može biti teži za čitanje i održavanje u poređenju sa iterativnim pristupom. Pored toga, postoji opasnost od grešaka prilikom razvoja rekurzivnih funkcija, uključujući beskonačne petlje i prevelike dubine rekurzije.

## Lokalni i globalni opseg vidljivosti

Za parametre i promenljive dodeljene u pozvanoj funciji kaze se da postoje u lokalnom opsegu vidljivosti funkcije. Za promenljive koje su dodeljene van svih funckija kaze se da postoje u globalnom opsegu vidljivosti. Promenljive moraju da budu jedno ili drugo, ne mogu da budu i lokalne i globalne. <br>

Opseg vidljivosti zamislite kao **kontejner za promenljive**. Kada je on unisten, sve vrednosti koje su sacuvane u promenljivim opsega vidljivosti su zaboravljene. Postoji samo jedan globalni opseg vidljivosti, koji je kreiran kada zapocnete pisanje programa. Kada se program zatvori, unisten je globalni opseg vidljivosti, a sve njegove promenljive su zaboravljene. U suprotnom, sledeceg puta kada pokrenete program, promenljive ce pamtiti svoje vrednosti koje su imale prilikom prethodnog pokretanja programa. <br>

Lokalni opseg vidljivosti je kreiran uvek kada je pozvana funckija. Sve promenljive koje su dodeljene u ovoj funkciji postoje i unutar lokalnog ospega vidljivosti. Kada se funkcija vrati, lokalni opseg vidljivosti je unisten, a ove promenljive su zaboravljene. Kada sledeceg puta pozovete ovu funkciju, lokalne promenljive nece pamtiti vrednosti koje su sacuvane u njima kada je funkcija prethodno bila pozvana. To ogranicavavrstu koda koji bi mogao da menja vrednosti promenljivih, sto moze biti korisno kada ispravljate greske u kodu. <br>

Opseg vidljivosti je vazan zbog nekoliko razloga:

1. Kod u globalnom opsegu vidljivosti ne moze da upotrebi lokalne promenljive.
2. Lokalni opseg vidljivosti moze da pristupi globalnim promenljivim.
3. Kod u lokalnom opsegu vidljivosti funkcije ne moze da upotrebi promenljive u bilo kojem drugom lokalnom opsegu vidljivosti.
4. Moze da se upotrebi isti naziv za razlicite promenljive ako se nalaze u razlicitim opsezima vidljivosti. Odnosno, moze postojati lokalna promenljiva pod nazivom *spam* i globalna promenljiva pod istim nazivom.



### 1. Lokalne promenljive ne mogu da se upotrebe u globalnom opsegu vidljivosti

```python
def spam():
    eggs=20
spam()
print(eggs)
```

Greska se javlja zato sto promenljiva eggs postoji samo u lokalnom opsegu vidljivosti koji je kreiran kada je pozvana funkcija spam().

### 2. Lokalni opsezi vidljivosti ne mogu da koriste promenljive u drugim lokalnim opsezima vidljivosti

```python
def spam():
    eggs=99
    bacon()
    print(eggs)

def bacon():
    hamp=101
    eggs=0
spam()
```

### 3. Globalne promenljive mogu da se citaju iz lokalnog opsega vidljivosti

```python
def spam():
    print(eggs)
eggs=42
spam()
print(eggs)
```

Posto ne postoji parametar pod nazivom eggs ili kod koji dodeljuje promenljivoj eggs vrednost u funkciji spam(), kada se eggs koristi u funkciji spam(), Python je smatra referencom za globalnu promenljivu eggs.

### Lokalne i globalne promenljive koje imaju isti naziv

Medjutim izbegavajte da koristite iste nazive zbog pracenja promenljivih u kodu!

```python
def spam():
    eggs='spam local' 
    print(eggs) #stampa spam local

def bacon():
    eggs='bacon local'
    print(eggs) #stampa 'bacon local'
    spam()
    print(eggs) #stampa 'bacon local'

eggs='global'
bacon()
print(eggs)
```

### Iskaz global

Ako treba da modifikujete globalnu promenljivu unutar funkcije, upotrebite iskaz *global*. Ako imate liniju kao sto je **global eggs** na vrhu funkcije, ona ,,govori" Pythonu: <br>
U ovoj funkciji eggs se odnosi na globalnu promenljivu, pa nemoj da kreiras lokalnu premenljivu pod istim imenom.


```python
def spam():
    global eggs
    eggs='spam'
    
eggs='global'
spam()
print(eggs)
```

Postoje cetiri pravila da odredite da li je promenljiva u lokalnom ili globalnom opsegu vidljivosti:

1. Ako je promenljiva upotrebljena u globalnom opsegu vidljivosti (odnosno, van svih funkcija), onda je uvek globalna.
2. Ako postoji iskaz **global** za odredjenu promenljivu u funkciji, to je globalna promenljiva.
3. U suprotnom, ako se promenljiva koristi u iskazu dodele u fukciji, ona je lokalna.
4. Ako promenljiva nije upotrebljena u iskazu dodele, to je globalna promenljiva.

```python
def spam():
    global eggs
    eggs='spam' # ovo je globalna prom
    
def bacon():
    eggs='bacon' # ovo je lokalna

def ham():
    print(eggs) # ovo je globalna

eggs=42 #ovo je globalna
spam()
print(eggs)
```

### Rukovanje izuzecima

Sada dobijanje greske ili izuzetka u Python programu znaci da ce ceo program pasti (nece se izvrsiti).
A to ne zelite da se to desi u programima koje kreirate. Umesto toga, zelite da program detektuje greske, da ih otkloni i zatim da nastavi da se pokrece. 

Primer program deljenje_sa_nulom.py

```python
def spam(divideBy):
    return 42 / divideBy
print(spam(2))
print(spam(12))
print(spam(0))
print(spam(1))
```

ZeroDivisionError greska se javlja kad god pokusate da podelite broj sa nulom. Zbog toga program se nije do kraja izvrsio. <br>
Greskama moze da se rukuje pomocu iskaza try i except. 
**Kod koji bi potencijalno mogao da ima gresku je postavljen u odredbu try. Izvrsenje programa se pomera na pocetak sledece odredbe except ako se desi greska. 
Nakon pokretanja tog koda, izvrsenje se nastavlja kao i normalno.**

```python
def spam(divideBy):
    try:
        return 42 / divideBy
    except ZeroDivisionError:
        print('greska, pogresan argument')
print(spam(2))
print(spam(12))
print(spam(0))
print(spam(1))
```

Razlog u primeru dole, zasto funcija spam(1) nikada nije izvrsena je cinjenica da, kada izvrsenje jednom preskoci na kod u odredbi except, nece se vratiti na odredbu try.

```python
# druga fora

def spam(divideBy):
     return 42 / divideBy

try:
    print(spam(2))
    print(spam(12))
    print(spam(0))
    print(spam(1))
except ZeroDivisionError:
    print('greska, pogresan argument')
```
