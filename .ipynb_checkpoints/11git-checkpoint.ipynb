## Bitni Git koncepti

Git ima tri glavna stanja u kojima se vai fajlovi mogu nai: **komitovano**, **modifikovano** i **stejdovano**. 

**Komitovano** znai da su podaci smeteni u lokalnoj bazi podataka na sigurnom. **Modifikovano** znai da ste promenili fajl ali da ga jo uvek niste komitovali u bazu podataka. **Stejdovano** znai da ste oznaili modifikovani fajl da se u svom trenutnom stanju ukljui u sledei komit snimka koji budete napravili.

Ovo nas vodi do tri glavne sekcije Git projekta: Git direktorijum, radni direktorijum i stejd.

**Git direktorijum** je mesto gde Git uva sve metapodatke i objektnu bazu podataka za va projekat. Ovo je najvaniji deo Gita, i to je ono to se kopira kada klonirate repozitorijum sa drugog raunara. Rezpozitorijum (ili repozitorij, eng. repository) je mesto gde se uva i odrava kolekcija podataka, obino vezana za softver, dokumentaciju ili druge vrste projekata. Repozitorijum moe biti lokalni direktorijum na raunaru ili serveru, ali se esto koristi i na internetu kako bi omoguio timsko radno okruenje, saradnju i praenje promena.

**Radni direktorijum** je jedan ekaut (checkout) jedne verzije projekta. Ovi fajlovi se dovlae iz kompresovane baze podataka iz Git direktorijuma i smetaju se na disk da biste ih koristili i modifikovali.

**Stejd** je fajl koji se obino nalazi u Git direktorijumu i uva informacije o tome ta e se ukljuiti u sledei komit. Ponekad se naziva indeks, ali se esto koristi i naziv stejd.

Osnovni tok rada sa Gitom ide nekako ovako: <br>
1.	Modifikujete fajlove u radnom direktorijumu. <br>
2.	Stejdujete fajlove, odnosno dodajete njihove snimke na stejd. <br>
3.	Komitujete, to znai da uzmete fajlove u onoj verziji u kojoj su bili kada ste ih dodali na stejd i te snimke trajno smetate u Git direktorijum. <br>

Ako se neka verzija fajla nalazi u Git direktorijumu, smatra se komitovanom. Ako je modifikovana i onda dodata na stejd, ona je stejdovana. A ako se promenila od ekauta ali nije stejdovana, onda je modifikovana.

Sada ako biste otili na repozitorijum na GitHub-u i osveili stranicu, videli biste da su vai fajlovi uspeno hostovani na GitHub-u. 


**"origin"** je skraenica koja se esto koristi kao ime za udaljeni repozitorijum sa kojim saraujete. Kada koristite komandu git clone kako biste preuzeli projekat sa udaljenog repozitorijuma, Git automatski postavlja "origin" kao podrazumevano ime za taj repozitorijum. Na primer, ako klonirate repozitorijum sa GitHub-a, "origin" e biti povezan sa URL-om tog GitHub repozitorijuma.

**"master"** je podrazumevana glavna grana u Gitu. Na njoj se obino nalazi najstarija i najstabilnija verzija koda. Meutim, vano je napomenuti da je u svetu razvoja softvera dolo do promene u nazivu glavne grane iz "master" u "main". Ova promena ima cilj eliminisanja potencijalno uvredljivog jezika i promociju inkluzivnosti. Dakle, umesto "master", mnogi projekti sada koriste "main" kao podrazumevani naziv glavne grane.


Kada se govori o operacijama kao to su git pull ili git push, "origin" i "master" (ili "main") se esto koriste zajedno da bi se naznailo da elite povui ili poslati promene na glavnu granu udaljenog repozitorijuma (npr. GitHub).

U Gitu moete imati vie grana osim glavne grane koja se obino naziva "master" ili "main". Ove dodatne grane omoguavaju vam rad na razliitim funkcionalnostima, ispravkama greaka ili eksperimentima nezavisno jedno od drugog. Na primer grana za testiranje odredjenih funkcionalnosti, ili ispravke gresaka, itd.

Sto se tice udaljenih repozitorijuma, u praksi, korisnici esto imaju samo jedan "origin" kao podrazumevani repozitorijum na koji alju svoje promene, ali imaju mogunost dodavanja dodatnih repozitorijuma sa razliitim aliasima kada sarauju sa vie timova ili repozitorijuma.

Povlaenje Najnovijih Izmena:

git pull kombinuje dve druge komande: **git fetch** i **git merge**.

git merge spaja promene sa odreene grane sa lokalnom granom.

Uobiajena upotreba ove komande je kada radite u timu i elite aurirati svoj lokalni rad sa najnovijim promenama koje su drugi lanovi tima ve poslali na master granu. **Preporuuje se da pre nego to uradite git pull, prvo izvrite git fetch kako biste videli ta se promenilo, a zatim, ako je sve u redu, izvrite git pull.**

**Detaljno objasnjenje 3. koraka**:
Dodavanje novih grana u Git repozitorijum je jednostavan proces. Evo kako moete dodati novu granu: <br>

1. Provera Trenutnih Grana:
Pre nego to dodate novu granu, dobro je proveriti trenutne grane u repozitorijumu. To moete uiniti koristei komandu: <br>
**git branch** <br>
2. Kreiranje Nove Grane: <br>
Za kreiranje nove grane koristite sledeu komandu: <br>
**git branch ime-nove-grane** <br>
3. Prebacivanje na Novu Granu: <br>
Da biste preli na novu granu, koristite: <br>
**git checkout ime-nove-grane** <br>
4. Dodavanje i Commitovanje Promena: <br>
Na ovoj novoj grani, moete raditi promene, dodati nove fajlove ili izmeniti postojee. Kada zavrite sa radom, dodajte i commit-ujte promene. <br>
5. Push Nove Grane na Udaljeni Repozitorijum: <br>
Ako elite deliti novu granu sa drugima, moete je push-ovati na udaljeni repozitorijum: <br>
**git push origin ime-nove-grane** <br>
6. Provera Trenutnih Grana (Opciono): <br>
Moete ponovo proveriti grane kako biste se uverili da ste preli na novu granu:<br>
**git branch** <br>


**NAPOMENA:Pre nego to zaponu rad na novoj grani, preporuuje se da kolaboratori povuku najnovije promene sa master grane ili glavne razvojne grane:** <br>
git checkout master <br>
git pull origin master <br>

**Spajanje Grana (Merge ili Pull Request):** <br>
Kada zavre sa radom na svojoj grani i ele integrisati promene u glavnu granu (master), mogu koristiti sledece komande: <br>
**git checkout master <br>
git merge nova-grana <br>
git commit -a -m "spajanje master  i nova-grana"** # komitovanje nastalih promena <br>
**git push origin master** #da bih sve gurnula nazad na udaljeni server <br>
git commit -a opcija stoji za "all" ili "autostage" i oznaava da elite automatski indeksirati (stage) sve promene koje su se desile u vaem radnom direktorijumu. Ovo ukljuuje izmene u ve **praenim fajlovima**.
Vano je napomenuti da -a ne indeksira nove fajlove koje Git ne prati. Ako ste dodali potpuno novi fajl koji Git ne prati, moraete ga eksplicitno dodati komandom git add.
