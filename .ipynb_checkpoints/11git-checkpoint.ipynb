## Bitni Git koncepti

Git ima tri glavna stanja u kojima se vaši fajlovi mogu naći: **komitovano**, **modifikovano** i **stejdžovano**. 

**Komitovano** znači da su podaci smešteni u lokalnoj bazi podataka na sigurnom. **Modifikovano** znači da ste promenili fajl ali da ga još uvek niste komitovali u bazu podataka. **Stejdžovano** znači da ste označili modifikovani fajl da se u svom trenutnom stanju uključi u sledeći komit snimka koji budete napravili.

Ovo nas vodi do tri glavne sekcije Git projekta: Git direktorijum, radni direktorijum i stejdž.

**Git direktorijum** je mesto gde Git čuva sve metapodatke i objektnu bazu podataka za vaš projekat. Ovo je najvažniji deo Gita, i to je ono što se kopira kada klonirate repozitorijum sa drugog računara. Rezpozitorijum (ili repozitorij, eng. repository) je mesto gde se čuva i održava kolekcija podataka, obično vezana za softver, dokumentaciju ili druge vrste projekata. Repozitorijum može biti lokalni direktorijum na računaru ili serveru, ali se često koristi i na internetu kako bi omogućio timsko radno okruženje, saradnju i praćenje promena.

**Radni direktorijum** je jedan čekaut (checkout) jedne verzije projekta. Ovi fajlovi se dovlače iz kompresovane baze podataka iz Git direktorijuma i smeštaju se na disk da biste ih koristili i modifikovali.

**Stejdž** je fajl koji se obično nalazi u Git direktorijumu i čuva informacije o tome šta će se uključiti u sledeći komit. Ponekad se naziva “indeks”, ali se često koristi i naziv stejdž.

Osnovni tok rada sa Gitom ide nekako ovako: <br>
1.	Modifikujete fajlove u radnom direktorijumu. <br>
2.	Stejdžujete fajlove, odnosno dodajete njihove snimke na stejdž. <br>
3.	Komitujete, što znači da uzmete fajlove u onoj verziji u kojoj su bili kada ste ih dodali na stejdž i te snimke trajno smeštate u Git direktorijum. <br>

Ako se neka verzija fajla nalazi u Git direktorijumu, smatra se komitovanom. Ako je modifikovana i onda dodata na stejdž, ona je stejdžovana. A ako se promenila od čekauta ali nije stejdžovana, onda je modifikovana.

**"origin"** je skraćenica koja se često koristi kao ime za udaljeni repozitorijum sa kojim sarađujete. Kada koristite komandu git clone kako biste preuzeli projekat sa udaljenog repozitorijuma, Git automatski postavlja "origin" kao podrazumevano ime za taj repozitorijum. Na primer, ako klonirate repozitorijum sa GitHub-a, "origin" će biti povezan sa URL-om tog GitHub repozitorijuma.

**"master"** je podrazumevana glavna grana u Gitu. Na njoj se obično nalazi najstarija i najstabilnija verzija koda. Međutim, važno je napomenuti da je u svetu razvoja softvera došlo do promene u nazivu glavne grane iz "master" u "main". Ova promena ima cilj eliminisanja potencijalno uvredljivog jezika i promociju inkluzivnosti. Dakle, umesto "master", mnogi projekti sada koriste "main" kao podrazumevani naziv glavne grane.


Kada se govori o operacijama kao što su git pull ili git push, "origin" i "master" (ili "main") se često koriste zajedno da bi se naznačilo da želite povući ili poslati promene na glavnu granu udaljenog repozitorijuma (npr. GitHub).

U Gitu možete imati više grana osim glavne grane koja se obično naziva "master" ili "main". Ove dodatne grane omogućavaju vam rad na različitim funkcionalnostima, ispravkama grešaka ili eksperimentima nezavisno jedno od drugog. Na primer grana za testiranje odredjenih funkcionalnosti, ili ispravke gresaka, itd.

Sto se tice udaljenih repozitorijuma, u praksi, korisnici često imaju samo jedan "origin" kao podrazumevani repozitorijum na koji šalju svoje promene, ali imaju mogućnost dodavanja dodatnih repozitorijuma sa različitim aliasima kada sarađuju sa više timova ili repozitorijuma.

**Detaljno objasnjenje 3. koraka**:
Dodavanje novih grana u Git repozitorijum je jednostavan proces. Evo kako možete dodati novu granu: <br>

1. Provera Trenutnih Grana:
Pre nego što dodate novu granu, dobro je proveriti trenutne grane u repozitorijumu. To možete učiniti koristeći komandu: <br>
**git branch** <br>
2. Kreiranje Nove Grane: <br>
Za kreiranje nove grane koristite sledeću komandu: <br>
**git branch ime-nove-grane** <br>
3. Prebacivanje na Novu Granu: <br>
Da biste prešli na novu granu, koristite: <br>
**git checkout ime-nove-grane** <br>
4. Dodavanje i Commitovanje Promena: <br>
Na ovoj novoj grani, možete raditi promene, dodati nove fajlove ili izmeniti postojeće. Kada završite sa radom, dodajte i commit-ujte promene. <br>
5. Push Nove Grane na Udaljeni Repozitorijum: <br>
Ako želite deliti novu granu sa drugima, možete je push-ovati na udaljeni repozitorijum: <br>
**git push origin ime-nove-grane** <br>
6. Provera Trenutnih Grana (Opciono): <br>
Možete ponovo proveriti grane kako biste se uverili da ste prešli na novu granu:<br>
**git branch** <br>


**NAPOMENA:Pre nego što započnu rad na novoj grani, preporučuje se da kolaboratori povuku najnovije promene sa master grane ili glavne razvojne grane:** <br>
git checkout master <br>
git pull origin master <br>

**Spajanje Grana (Merge ili Pull Request):** <br>
Kada završe sa radom na svojoj grani i žele integrisati promene u glavnu granu (master), mogu koristiti sledece komande: <br>
**git checkout master <br>
git merge nova-grana <br>
git commit -a -m "spajanje master  i nova-grana"** # komitovanje nastalih promena <br>
**git push origin master** #da bih sve gurnula nazad na udaljeni server <br>
git commit -a opcija stoji za "all" ili "autostage" i označava da želite automatski indeksirati (stage) sve promene koje su se desile u vašem radnom direktorijumu. Ovo uključuje izmene u već **praćenim fajlovima**.
Važno je napomenuti da -a ne indeksira nove fajlove koje Git ne prati. Ako ste dodali potpuno novi fajl koji Git ne prati, moraćete ga eksplicitno dodati komandom git add.
