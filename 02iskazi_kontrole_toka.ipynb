### Bulovi (logicki) operatori
Bulovi operatori and,or, i not se koriste za uporedjivanje Bulovih vrednosti. Kao i operatori poredjenja, oni svode ove izraze na Bulovu vrednost.

#### Binarni Bulovi operatori
Operatori *and* i *or* uvek uzimaju dve Bulove vrednosti (ili izraza), pa se, stoga, smatraju binarnim operatorima.

Operator **and**

| Izraz |    Svodi na |
|-----------|-------------|
|True and True         | True   |
|True and False         | False   |
|False and True         | False   |
|False and False        | False   |

Operator **or**

| Izraz |    Svodi na |
|-----------|-------------|
|True or True         | True   |
|True or False         | True   |
|False or True         | True   |
|False or False        | False   |

#### Operator not
Za razliku od operatora and i or, operator not deluje samo na jednu Bulovu vrednost (ili izraz). On jednostavno svodi vrednost na suprotnu Bulovu vrednost.



```python
(4 < 5) and (5 < 6)
```

```python
2 + 2 == 4 and not 2 + 2 == 5 and 2*2 == 2 + 2
```

> Bulovi operatori imaju redosled operacija isti kao i matematicki operatori. Nakon sto matematicki operatori i operatori poredjenja izvrse izracunavanje, Python prvo koristi operatore not, zatim operatore and, pa operatore or.

### Iskazi IF

Najcesce upotrebljivana vrsta iskaza kontrole toka je iskaz if. Odredba tog iskaza (odnosno bloka), koji prati iskaz if, ce se izvrsiti ako je uskov iskaza tacan (True). Odredba se izostavlja ako je uslov netacan (False).

U Python-u iskaz if se sastoji od sledeceg:
1. kljucne reci if
2. uslova (odnosno, izraza koji se svodi na True ili False)
3. dvotacke
4. Zapocinje u sledecoj liniji uvucenog bloka koda (naziva se odredba if)



```python
name='Ana'
if name=='Ana':
    print('Cao Ana')
print('zavrsen if uslov!')
```

#### Iskaz else

Odredba if moze opciono da bude pracena iskazom else. Odredba else se izvrsava samo kada je uslov iskaza if netacan (False). Iskaz else nema uslov uz sebe.

```python
name='Jovana'
if name=='Ana':
    print('Cao Ana')
else:
    print('Cao strance')
print('zavrsen if uslov!')
```

#### Iskaz elif

Iako ce biti izvrsena samo jedna od if ili else odredbi, mozda cete zeleti da se izvrsi jedna od mnogih mogucih odredbi. Iskaz elif je iskaz else if, koji uvek prati iskaz if ili drugi else if uskaz. Iskaz elif obezbedjuje drugi uslov koji se proverava samo ako je bilo koji od prethodnih uslova bio netacan (False). 

```python
name='Ana'
age=1000

if name=='Alisa':
    print('Cao Alisa')
elif age < 12:
    print('Ti nisi Alisa, dete')
elif age > 2000:
    print('Za razliku od tebe, Alisa nije vampir')
elif age > 100:
    print('Bako, ti nisi Alisa')
print('Zavrsila')
```

### Petlje while

Mozete da ucinite da se blok koda izvrsava iznova pomocu iskaza while. Kod u odredbi while ce biti izvrsavan sve dok je uslov iskaza while True. U kodu iskaz while uvek se sastoji od sledeceg:
1. kljucne reci while
2. uslova (odnosno izraza koji se svodi na vrednost True ili False)
3. dvotacke
4. Pocinje u sledecoj liniji uvucenog bloka koda

Vidite da je iskaz *while* slican iskazu *if*. Razlika je u njihovom ,,ponasanju''. Na kraju odredbe *if* izvrsavanje programa se nastavlja nakon iskaza *if*. 
Medjutim, na kraju odredbe *while* izvrsenje programa se vraca na pocetak iskaza *while*. Odredba *while* se cesto naziva **petlja while** ili samo petlja.

```python
spam = 0
while spam < 5:
    print('Hello world')
    spam = spam + 1
```

##### Dosadna petlja while

```python
name=''
while name != 'your name':
    print('Please type your name')
    name=input()
print('Thank you')
```

### Iskazi break

Postoji precica za ranije prekidanja izvrsavanja programa odredbe petlje while. Ako izvrsavanje dodje do iskaza break, odmah ce napustiti odredbu petlje while. U kodu se iskaz break sastoji samo od kljucne reci break. 

```python
while True:
    print('Please type your name')
    name=input()
    if name == 'your name':
        break
print('Thank you')
```

### Iskazi continue

Kao i iskazi break, iskazi continue se koriste unutar petlji. Kada izvrsenje programa dodje do iskaza continue, odmah se vraca na pocetak petlje i ponovo izracunava uslov petlje. <br>
**To se desava i kada izvrsenje stigne do kraja petlje**

```python
while True:
    print('Ko si ti?')
    name = input()
    if name != 'Irida':
        continue
    print('Zdravo Irida. Koja je tvoja lozinka(To je riba)?')
    password = input()
    if password == 'losos':
        break
print('Pristup omogucen')
```

Postoje neke vrednosti u drugim vrstama podataka koje ce uslovi smatrati ekvivalentom za True i False. Kada su upotrebljene vrednosti u uslovima, 0, 0.0 i '' (prazan string), smatraju se neistinite (False), dok se sve druge vrednosti smatraju kao istinite (True).


```python
name = ''
while not name:
    print('Ukucaj svoje ime')
    name = input()
    print('Koliko ces gostiju imati')
    numOfGuests = int(input())
    if numOfGuests:
        #ako broj gostiju nije 0, onda se uslov smatra True
        print('Pazi da imas dovoljno mesta za sve goste')
    print('Zavrseno')
```

### Petlja for i funkcija range()

Petlja while nastavlja da kruzi, dok usliv ne bude True (sto je i razlog njenog naziva), ali sta se desava ako zelite da izvrsite blok koda samo odredjeni broj puta? <br>
Da bi blok koda bio izvrsen odredjeni broj puta, upotrebite iskaz petlje for i funckiju range(). <br>


```python
print('My name is')
for i in range(5):
    print('Jimmy Five Times (' + str(i) + ')')
```

### Ranije zavrsavanje programa pomocu funkcije sys.exit()

Poslednji koncept kontrole toka koji treba da opisemo je nacin kako se zavrsava program. Ovo se uvek desava kada izvrsenje programa stigne do kraja instrukcija. 

Medjutim,mozete da izazovete zavrsavanje ili zatvaranje programa pozivanjem funckije sys.exit(). Posto se ova funkcija nalazi u modulu sys, pre nego sto je upotrebite u programu, treba da importujete modul **sys**.


```python
import sys

while True:
    print('Ukucaj exit kako bi izasao iz programa')
    response = input()
    if response == 'exit':
        sys.exit()
    print('Ukucali ste ' + response + '.')
```
